```python
#write a function to count the number of entries in series
def count_values(series):
    return series.count()

#write a function to find the maximum number in series
def max_num(series):
    maximum = series[0]
    for i in series:
        if(i>maximum):
            maximum =i
    return maximum

#write a function to find the minimum number in series
def min_num (series):
    minimum = series[0]
    for i in series:
        if(i<minimum):
            minimum = i
    return minimum

#write a functiont to find the mean of series
def mean_value(series):
    sum = 0
    for i in series:
        sum = i + sum
        mean = sum/count_values(series)
        rounded_mean = round(mean, 4)
    return rounded_mean

def median(series):
    count = count_values(series)
    sorted_series = sorted(series)
    if (count % 2 != 0):
        return sorted_series[count//2]
    else:
        e1 = sorted_series[count//2]
        e2 = sorted_series[(count//2) - 1]
        return (e1 + e2)/2
        
    
    
```

```python
#Quantile Function

arr = [1,3,5,13,7,9,15,18,20,25]
p = 25
sorted_arr = sorted(arr)

R1 = (p/100)*(len(sorted_arr) + 1)

# if R1%10 == 0:
#     print(sorted_arr[R1])
# else:
#     R1 = int(R1)
#     R2 = R1 + 1
#     print((p/100)*(sorted_arr[R1] - sorted_arr[R2]))

if type(R1) is int:
    print(sorted_arr[R1])
if type(R1) is float:
    #Then use formula
#     Q is the 25th percentile.
#     L is the value at the lower position (3).
#     N is the total number of data points (10).
#     F is the integer part of the index of the 25th percentile (2).
#     U is the value at the upper position (5).
    L = sorted_arr[int(R1)]
    U = sorted_arr[(int(R1)+1)]
    N = len(arr)
    F = int(R1)
    Q = L - (N/4 - F) * (U - L)
    
    print(Q)
```
